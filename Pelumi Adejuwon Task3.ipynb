```python
merged_df1 = pd.merge(pizzadata_df1,pizzadata_df2, on='order_id',how='inner')
merged_df1
```

```python
merged_df2 =pd.merge(pizzadata_df3,pizzadata_df4,on ='pizza_type_id', how ='inner')
merged_df2
```

```python
dataset_df =pd.merge(merged_df1,merged_df2, on='pizza_id', how ='inner')
dataset_df
```

```python
#to combine date and time in a single column

dataset_df['datetime']=pd.to_datetime(dataset_df['date'] +' '+dataset_df['time'])
```

```python
dataset_df.duplicated().sum()    #to check duplicates
```

```python
dataset_df["revenue"] =dataset_df["price"]*dataset_df["quantity"]     #to create a revenue column
```

```python
dataset_df.head()
```

```python
#to find total revenue



total_revenue=dataset_df["revenue"].sum()
```

```python
print("total_revenue",total_revenue)
```

```python
#to find the total quantity sold

total_quantity= dataset_df["quantity"].sum()

print("total_quantity",total_quantity)
```

```python
# to find the total orders


total_orders =dataset_df["order_id"].nunique()
print ("total_orders",total_orders)
```

```python
#to find the no of pizza types being sold

pizza_types =dataset_df["pizza_Type"].unique()
print ("pizza_types",pizza_types)
```

```python
num_pizza_types =dataset_df["pizza_Type"].nunique()
print ("total number of pizza types",num_pizza_types)
```

```python
#average price of the pizzas

avg_price = dataset_df["price"].mean()
print ("Average pizza price:",avg_price)


```

```python
#to find the peak hours of sales


dataset_df["hour"]=dataset_df["time"]
```

```python
orders_per_hour=dataset_df["hour"].value_counts().sort_index()
print(orders_per_hour)
```

```python
#now to find the peak hour

peak_hour=orders_per_hour.idxmax()
print ("peaksales hour:", peak_hour)
```

```python
#to find the sales made on each day of the week and which day had the most sales

#first,extract the day of the week
dataset_df['day_of_the_week']= dataset_df['datetime'].dt.day_name()
```

```python
#total sales by weekday

sales_per_day=dataset_df.groupby('day_of_the_week')['revenue'].sum()
```

```python
sales_per_day= sales_per_day.sort_values(ascending=False)
print (sales_per_day)
```

```python
#day with the most sales

top_day=sales_per_day.idxmax()
print(f"The day with the most sales is :{top_day}")
```

```python
#top 5 best selling pizzas


#first find the total revenue per pizza
pizza_revenue= dataset_df.groupby('full_name')['revenue'].sum()
```

```python
top_5_pizzas= pizza_revenue.sort_values(ascending=False).head(5)
print(top_5_pizzas)
```

### Sales made in each month and trends noticeable ###

```python
#extract month and year first by creating a month year column

dataset_df['month_year']= dataset_df['datetime'].dt.strftime('%Y-%m')
monthly_sales=dataset_df.groupby('month_year',as_index=False)['revenue'].sum()
monthly_sales= monthly_sales.sort_values('month_year')

print(monthly_sales)
```

```python
dataset_df.head()
```

```python
print(dataset_df['datetime'].head())
print(dataset_df['datetime'].dtype)
```

```python
plt.plot(monthly_sales['month_year'],monthly_sales['revenue'])
plt.title("Monthly pizza revenue trend")
plt.xlabel('month_year')
plt.ylabel('revenue')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
